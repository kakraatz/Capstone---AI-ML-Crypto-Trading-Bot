### Imports

```python
import json
import requests
```

### Connecting to Binance.US

```python
symbol = "BTCUSD"
interval = "1d"
candles = requests.get(f"https://api.binance.us/api/v3/klines?symbol={symbol}&interval={interval}")
candles_json = []
if candles.status_code == 200:
    candles_json = json.loads(candles.text)
    print(candles_json)
```
